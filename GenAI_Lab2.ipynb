```python
from scipy.optimize import linprog
import numpy as np
import math as mt
```

```python
obj1 = [-6, -5]
A_eq1 = [[1, 1], [3,2]]
B_eq1 = [5,12]
bnd = [(0, float("inf")), (0, float("inf"))]
opt = linprog(c=obj1, A_eq=A_eq1, b_eq=B_eq1, bounds=bnd)
choc = np.round(opt.x)
print(f"No. of A :{int(choc[0])}")
print(f"No. of B :{int(choc[1])}")
```
